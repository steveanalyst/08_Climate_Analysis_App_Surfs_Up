# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
measure_first = session.query(Measurement).first()
measure_first
measure_first.__dict__
```

```python
# Query to get the max or latest date
max_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
max_date = max_date[0]
max_date
```

```python
# Calculate the last 12 month beginning date
year_ago_date = dt.datetime.strptime(max_date, "%Y-%m-%d") - dt.timedelta(days=365)
year_ago_date.strftime('%Y-%m-%d')
```

```python
# Perform a query to retrieve the data and precipitation scores
prcp_12_month = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= year_ago_date).\
       group_by(Measurement.date).all()
prcp_12_month
```

```python
# Save the query results as a Pandas DataFrame
precipitation_df = pd.DataFrame(prcp_12_month)
precipitation_df.head()
```

```python
#set the index to the date column
prcpt_df = precipitation_df.set_index("date")
prcpt_df.head()
```

```python
# Sort the dataframe by date
prcpt_df.sort_values(by=['date'])
prcpt_df.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
prcpt_df.plot(figsize = (10,8), rot = 45, use_index = True, legend=True)
# Set plot title and show
plt.title('Hawaii 2016-08-23 to 2017-08-23 Precipitation Data')
plt.ylabel('Precipation')
plt.savefig("Images/HawaiiPrecipitationData.png")
plt.show()
```

```python
# Design a query to show how many stations are available in this dataset?
Stations = session.query(Measurement).group_by(Measurement.station).count()
Stations
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

active_stations = session.query(Measurement.station,func.count(Measurement.station)).\
group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()

active_stations
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
most_active = active_stations[0][0]
most_active_temps = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs),
                                  func.avg(Measurement.tobs)).filter(Measurement.station == most_active).all()
most_active_temps
```

```python
# Choose the station with the highest number of temperature observations.
max_count_station = session.query(Measurement.station, func.count(Measurement.tobs)).\
group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).first()
max_count_station= max_count_station[0]
max_count_station
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
trip_temp=calc_temps('2017-08-15', '2017-08-23')
trip_temp
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

start_date="2017-08-15"
end_date="2017-08-23"
trip_temp = session.query(Measurement.date, Measurement.tobs).filter(Measurement.date >= start_date,\
         Measurement.date <= end_date).all()
trip_temp = pd.DataFrame(trip_temp)
max_temp = trip_temp["tobs"].max()
min_temp = trip_temp["tobs"].min()
mean_temp = trip_temp["tobs"].mean()
plt.figure(figsize=(3,6))
plt.ylim(0,100)     # set the ylim to bottom, top
plt.bar(1,mean_temp, yerr= max_temp - min_temp, tick_label="",color='sandybrown',alpha=0.75)
plt.title('Trip Avg Temp', fontsize=25, color='black')
plt.ylabel('Temp (F)', fontsize = 18,color='black')
plt.tight_layout()
plt.savefig("Images/TripAvgTemp.png")
plt.show()
```

```python
# Calculate the rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

def prcp(start_date, end_date):
    
    """
    
    rainfall data per weather station for given trip dates
    Using previous year's matching dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        Tuple list for:
        precipitation amount (descending order), 
        station, 
        name, 
        latitude, 
        longitude, 
        elevation 
        
    """
    
    list = [Measurement.station, 
           Station.name, 
           Station.latitude, 
           Station.longitude, 
           Station.elevation, 
           Measurement.prcp]
    
    return session.query(*list).\
            filter(Measurement.station == Station.station).\
            filter(Measurement.date >= start_date).\
            filter(Measurement.date <= end_date).\
            group_by(Measurement.station).\
            order_by(Measurement.prcp.desc()).all()

print(prcp('2017-08-15','2017-08-23'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
#################################################################
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`
#################################################################

# Set the start and end date of the trip
start_date = '2017-08-15'
end_date ='2017-08-23'

# Use the start and end date to create a range of dates
dates = session.query(Measurement.date).\
        filter(Measurement.date >= start_date).\
        filter(Measurement.date <= end_date).\
        group_by(Measurement.date).all()
dates
```

```python
trip_dates = [x[0] for x in dates]
trip_dates
```

```python
# Strip off the year and save a list of %m-%d strings
trip_dates_md= [x[5:] for x in trip_dates]
trip_dates_md
```

```python
# Loop through the list of %m-%d strings and calculate the normals for each date
start_md = trip_dates_md[0]
end_md = trip_dates_md[-1]

daily_normals = [session.query(func.min(Measurement.tobs),
                       func.avg(Measurement.tobs),
                       func.max(Measurement.tobs)).\
                       filter(func.strftime("%m-%d", Measurement.date) >= start_md).\
                       filter(func.strftime("%m-%d", Measurement.date) <= end_md).\
                       group_by(func.strftime("%m-%d", Measurement.date)).all()]

daily_normals = daily_normals[0]
daily_normals
```

```python
# Load the previous query results into a Pandas DataFrame 
daily_normals_df= pd.DataFrame(daily_normals,columns=['tmin', 'tavg', 'tmax'])
daily_normals_df
```

```python
# Add trip_dates to Pandas DataFrame
daily_normals_df['date']= trip_dates
daily_normals_df
daily_normals_df.dtypes
```

```python
# Change datatype from element to datetime object on Pandas
daily_normals_df['date'] = pd.to_datetime(daily_normals_df['date'], format='%Y-%m-%d')
daily_normals_df.dtypes
daily_normals_df
```

```python
# Set index to date
daily_normals_df.set_index('date',inplace=True)
daily_normals_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
daily_normals_df.plot(kind='area', alpha=.2, stacked=False, x_compat=True, figsize=(9,6),title="Daily Normals for Given Trip Dates")
plt.ylabel("Daily Normals in F")
plt.xlabel("Trip Dates")
plt.tight_layout()
plt.savefig("Images/DailyNormalsGivenDates.png")
plt.show()
```
